# Analysis of Global Crop Yields and Climate Sensitivity

# Introduction

Global agricultural systems face mounting pressures from shifting climate patterns. While overall crop yields have steadily increased over the past several decades due to technological advancements, these macro-trends often mask underlying vulnerabilities to environmental shocks. This project studies a comprehensive dataset of global crop yields, mapping them against annual average temperatures, rainfall, and pesticide usage across multiple decades. 

In this report, we will isolate human-driven technological growth from natural climate signals to identify which crops and regions are most sensitive to environmental fluctuations. To achieve this, we will apply polynomial detrending, calculate non-linear Mutual Information ($I(X;Y)$) metrics, and utilize Principal Component Analysis (PCA) paired with K-Means clustering to group crops by their climate risk profiles. Finally, we will project these findings onto an interactive geospatial map.

Dataset : https://www.kaggle.com/datasets/mrigaankjaswal/crop-yield-prediction-dataset/data

### 1. Methodology: Isolating the Climate Signal

A fundamental challenge in agricultural data science is the confounding effect of time. Over the last 30 years, base crop yields have naturally increased due to steady technological advancements, such as improved genetics, better machinery, and optimized fertilizer use. If we naively correlate raw yield with annual climate data, the results are artificially skewed by this linear progression of time.

To resolve this, we must first establish a baseline. The visualizations below demonstrate our detrending methodology. By fitting a Degree-2 Polynomial to the raw yield time-series, we capture the "technological trend." We then subtract this trend from the raw data. The resulting residuals—our "Detrended Anomaly" ($\epsilon$)—represent the pure environmental shocks caused by year-to-year weather variations. This detrended metric serves as the foundation for our subsequent climate sensitivity analysis.

```python
fig_detrend = plot_detrending_example(df, country='India', crop='Wheat')
```

### 2. Marginal and Joint Distributions

Before calculating strict correlation matrices, it is crucial to examine the raw distributions of our environmental factors to ensure our statistical approaches are appropriate. 

The joint density plot below illustrates the relationship between average annual rainfall and the logarithmic transformation of yield, $\log(1 + \hat{Y})$. The marginal histograms reveal that both variables exhibit distinct, non-normal distributions with significant skewness. Furthermore, the scatter plot highlights heavy clustering rather than a uniform linear spread, indicating that standard parametric tests might fail to capture the true underlying dynamics.

```python
fig_joint = plot_joint_distribution(df, crop_name='Wheat')
```

### 3. The Pesticide Paradox (Why we use Mutual Information)

Standard linear correlation (Pearson's $\rho_{XY}$) assumes a straight-line relationship between variables. However, ecological and chemical systems rarely scale infinitely. 

The plot below visualizes the "Pesticide Paradox." As pesticide usage increases, yield does not increase linearly; instead, it exhibits a logarithmic "diminishing returns" effect. Eventually, chemical saturation is reached, and additional pesticides provide negligible yield benefits (and may even cause ecological harm). Because of these inherent non-linearities, using Pearson correlation would misrepresent the data. Therefore, we utilize **Mutual Information**, $I(X;Y)$, to calculate our sensitivity metrics, as it is capable of capturing both linear and highly non-linear dependencies.

```python
fig_pest = plot_pesticide_paradox(df, crop='Maize')
```

### 4. Principal Component Analysis (PCA)

Having calculated the Mutual Information sensitivities for Temperature, Rainfall, and Pesticide usage across all crops, we now face a high-dimensional problem. To identify global patterns, we pass these sensitivity vectors through Principal Component Analysis (PCA) and apply K-Means clustering.

The resulting biplot projects this complex matrix into a 2D space. The groupings here are highly revealing:
* **Proximity** indicates crops that share similar physiological responses to environmental stressors.
* **Marker Size** (Pesticide Volume) reveals that highly chemical-dependent crops often cluster together, suggesting that intensive farming practices may alter a crop's natural climate resilience profile.
* **Color** (Continent) shows regional biases, highlighting how the same crop might behave differently depending on its geographic and economic ecosystem.

```python
fig_pca = plot_pca_biplot(pca_df)
```

### 8. Limitations and Future Scope

While this framework provides a robust macro-level view of agricultural climate sensitivity, it is important to acknowledge the limitations of the dataset:
* **Temporal Resolution:** This data relies on annual averages. Agriculture is highly sensitive to micro-seasonal events (e.g., a two-week drought during the critical pollination phase), which annual data can obscure.
* **Missing Variables:** The dataset lacks soil quality indices, localized irrigation infrastructure metrics, and granular breakdowns of specific fertilizer/pesticide types, all of which heavily influence yield resilience.

**Future Work:** Future iterations of this project would involve integrating high-frequency, monthly weather data and utilizing time-series forecasting models (like ARIMA or LSTMs) to predict future yield anomalies based on projected climate change scenarios.
